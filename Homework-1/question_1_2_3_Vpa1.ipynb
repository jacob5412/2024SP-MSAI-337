### 1.Tokenization Strategy:

**NLTK**: Uses a more traditional method, splitting the text based on spaces and punctuation. NLTK's approach can be seen as relatively straightforward and rule-based, where tokens often correspond to clear-cut word divisions and punctuation marks.

**GPT-2**: Employs a more complex strategy using Byte-Pair Encoding (BPE), which creates tokens based on frequently occurring substrings in the training data. This results in tokens that might include partial words, subwords, combinations of characters, and even whitespace (represented as 'Ġ') to denote spaces following other tokens.

### 2.Handling Special Characters and Unrecognized Tokens:

**NLTK**: Directly tokenizes text as it appears, without specific handling for characters or strings not recognized. Special or foreign characters are tokenized based on their separation by spaces or punctuation.

**GPT-2**: Uses tokens like 'Ġ<' and 'unk' to handle unrecognized characters or sequences not common in its training set. The 'unk' token indicates an 'unknown' sequence that the tokenizer could not match to its vocabulary.

### 3.Granularity:

**NLTK**: Generally maintains a coarse granularity, where each token is often a whole word or a punctuation mark.

**GPT-2**: Exhibits finer granularity, breaking down words into smaller parts. This can be seen in how it tokenizes names, unrecognized sequences, or sequences involving non-English characters, which might be broken down into several sub-tokens to capture the underlying structure better.

### 4.Handling of Whitespace:

**NLTK**: Does not explicitly tokenize whitespaces; spaces are used as delimiters to separate tokens.

**GPT-2**: Includes an explicit representation of spaces in certain contexts (denoted by 'Ġ'), which is particularly useful in differentiating words in sequences where whitespace is significant for interpretation or generation tasks.

```python
from utils.ngram_utils import get_ngrams_from_file, ngram_frequencies, calculate_perplexity
# Define file paths
train_file_path = 'data/wiki2.train.txt'
test_file_path = 'data/wiki2.test.txt'
```

```python
def perplexity_looper(n):
    # Load and process training data
    train_ngrams = get_ngrams_from_file(train_file_path, n)  
    train_freqs = ngram_frequencies(train_ngrams)
    total_train_ngrams = sum(train_freqs.values())
    vocabulary = set(word for ngram in train_ngrams for word in ngram.split())
    vocabulary_size = len(vocabulary)

    # Load and process test data
    test_ngrams = get_ngrams_from_file(test_file_path, n)


    # Calculate perplexity
    perplexity = calculate_perplexity(test_ngrams, train_freqs, total_train_ngrams, vocabulary_size)
    return perplexity
```

```python
for i in range(1,7):
    perplexity = perplexity_looper(i)
    print("Perplexity of the test set for " +  str(i) +f"-grams: {perplexity}\n")
```
